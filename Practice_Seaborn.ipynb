```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day')
plt.title("Total BILL")
print(plt)
```

```python
sns.swarmplot(data=tips,x='total_bill',y='tip',hue='day')
plt.title("Total BILL")
print(plt)
```

```python
sns.stripplot(data=tips,y='total_bill',x='sex',hue='sex')
plt.title("Total BILL")
print(plt)
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
plt.title("Total bill")
plt.show()
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
plt.title("Total bill")
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box',color='black',hue='sex')
plt.title("Total bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.displot(data=iris, x='sepal_length', kde=True, col='species')
g.fig.suptitle("Distribution Plot of Sepal Length by Species", ha='center')
g.fig.subplots_adjust(top=0.85)
plt.show()
```
